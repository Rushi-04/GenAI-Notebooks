```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain.agents import create_agent
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq(model="openai/gpt-oss-120b")
```

```python
search = GoogleSerperAPIWrapper()
```

```python
search.run("Narendra Modi's full name ?")
```

```python
agent = create_agent(
    model=llm,
    tools=[search.run],
    system_prompt="You are a agent and can search for any question on google."
)
```

```python
question = "What was the weather yesterday at Washim, Maharashtra ?"
```

```python
reponse = agent.invoke({"messages": [{"role": "user", "content": question}]})
```

```python
print(reponse["messages"][-1].content)
```
