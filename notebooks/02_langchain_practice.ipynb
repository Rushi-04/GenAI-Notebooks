### Prompt Templates - Dynamic Prompts

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
prompts = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are translator and translate input in {language}"},
    {"role": "user", "content": "{query}"}
])
```

```python
final_prompts = prompts.format_messages(language = "Marathi",
                                        query = "It's not over until I win, and if it is failure then it's not the end")
```

```python
res = llm.invoke(final_prompts)
```

```python
print(res.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

#### Other method

```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
outParser = StrOutputParser()
```

```python
prompts = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are translator and translate input in {language}"},
    {"role": "user", "content": "{query}"}
])
```

```python
##### Custom chain part
def convert_to_upper(result:str):
    return result.upper()
```

```python
chains = prompts | llm | outParser | convert_to_upper
```

```python
res = chains.invoke({"language": "Hinglish", "query": "Hello Pratik bhau, welcome to langchain class"})
```

```python
print(res)
```
