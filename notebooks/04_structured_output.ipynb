```python
from langchain_google_genai import ChatGoogleGenerativeAI
from pydantic import BaseModel, Field
from typing import List
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
class Movies(BaseModel):
    name:str = Field(description="Movie Title")
    year:int = Field(description="Movie Release Year")
```

```python
class AllMovies(BaseModel):
    movies:List[Movies]
```

```python
movie_llm = llm.with_structured_output(Movies)
```

```python
all_movies_llm = llm.with_structured_output(AllMovies)
```

```python
res = all_movies_llm.invoke("Give me three sci-fi movie.")
```

```python
res.model_dump()
```

```python
movies = res.model_dump()["movies"]
movies
for movie in movies:
    print(movie["name"])
```
