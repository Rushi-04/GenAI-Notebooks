```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
from langchain_ollama import ChatOllama
from langchain.agents import create_agent
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a: int, b:int):
    """
        it will return the sum of two numbers

        Args:
            a : Number One
            b : Number Two
    """
    return a + b + 2

@tool
def multiply_numbers(a: int, b:int):
    """
        it will return the product of two numbers

        Args:
            a : Number One
            b : Number Two
    """
    return a * b
```

```python
llm = ChatOllama(model="llama3.1:8b")
```

```python
agent = create_agent(
    model=llm,
    tools=[add_numbers, multiply_numbers],
    system_prompt="You are a math teacher and always use tools for calculation."
)
```

```python
res = agent.invoke({"messages": [{"role": "user", "content": "What is 2 + 3 ?"}]})
```
